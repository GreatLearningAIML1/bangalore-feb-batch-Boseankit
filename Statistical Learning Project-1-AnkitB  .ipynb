```python
# Load the necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import operator

mydata = pd.read_csv('Laliga.csv')
```

```python
#1.Read the data set 
mydata.head(n= 20)
```

```python
#replace dashes with 0 
mydata = mydata.replace('-', 0)
```

```python
#moving the Column as column decription
mydata.columns = mydata.iloc[0]
mydata = mydata[1:]
```

```python
mydata.dropna(inplace = True) 
```

```python
mydata.head(20)
```

```python
#converting to Int so Arthematic operations can be done

mydata['Points'] = mydata['Points'].astype(int)
mydata['Seasons'] = mydata['Seasons'].astype(int)
mydata['GamesPlayed'] = mydata['GamesPlayed'].astype(int)
mydata['GamesWon'] = mydata['GamesWon'].astype(int)
mydata['GamesDrawn'] = mydata['GamesDrawn'].astype(int)
mydata['GamesLost'] = mydata['GamesLost'].astype(int)
mydata['GoalsFor'] = mydata['GoalsFor'].astype(int)
mydata['GoalsAgainst'] = mydata['GoalsAgainst'].astype(int)
mydata['Champion'] = mydata['Champion'].astype(int)
mydata['Runner-up'] = mydata['Runner-up'].astype(int)
mydata['Third'] = mydata['Third'].astype(int)
mydata['Fourth'] = mydata['Fourth'].astype(int)
mydata['Fifth'] = mydata['Fifth'].astype(int)
mydata['Sixth'] = mydata['Sixth'].astype(int)
mydata['T'] = mydata['T'].astype(int)
mydata['BestPosition'] = mydata['BestPosition'].astype(int)


mydata.info()
```

```python
''''4  Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
Using the same function, find the team which has maximum and minimum goal difference. 
Goal_diff_count = GoalsFor - GoalsAgainst'''''
def Goal_diff_count(a) : 
    mydataF = a
    mydataF["Goal_diff_count"] = mydataF["GoalsFor"] - mydataF["GoalsAgainst"]
    mydataF = mydataF.sort_values('Goal_diff_count',ascending = False)
    print("Team details with Max Goal diff is below ")
    print(mydataF.loc[1])
    print("                                 ")
    print("Team details with Min Goal diff is below")
    print(mydataF.loc[61])

Goal_diff_count(mydata)
```
